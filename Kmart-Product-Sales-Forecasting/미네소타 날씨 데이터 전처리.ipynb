```python
data = pd.read_csv("M_weather.csv")
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.replace('T', 0, inplace=True)
```

```python
data['Precipitation (inches)'] = data['Precipitation (inches)'].astype('float')
data['Snow (inches)'] = data['Snow (inches)'].astype('float')
data['Snow Depth (inches)'] = data['Snow Depth (inches)'].astype('float')
```
