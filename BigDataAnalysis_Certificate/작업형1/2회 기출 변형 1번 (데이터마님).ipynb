# 1-2
1-1에서 사용한 데이터에서 RM 중앙값으로 해당 컬럼의 결측치를 대체하라

그리고 해당 컬럼의 결측치 대치 전후의 표준편차 차이의 절댓값을 소숫점 이하 3째자리 까지 구하라

```python
data1['RM'].isnull().sum()
```

```python
data1['RM'].median()
```

```python
# 이전 표준편차
pre_std = data1['RM'].std()
pre_std
```

```python
data1['RM'].fillna(data1['RM'].median(), inplace=True)
```

```python
data1['RM'].isnull().sum()
```

```python
after_std = data1['RM'].std()
after_std
```

```python
round(abs(pre_std-after_std), 3)
```

# 1-3

주어진 Dataset의 DIS 평균으로 부터 1.5 * 표준편차를 벗어나는 영역을 이상치라고 판단하고

DIS 컬럼의 이상치들의 합을 구하여라.

```python
mean = data1['DIS'].mean()
mean
```

```python
std = data1['DIS'].std()
std
```

```python
data1['DIS'].sum()
```

```python
data1[(data1['DIS']> 1.5*std+mean) | (data1['DIS']< mean-1.5*std)].DIS.sum()
```
