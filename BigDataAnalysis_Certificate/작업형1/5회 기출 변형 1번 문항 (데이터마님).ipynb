```python
data2['height(m)'] = data2['height(cm)'] / 100
```

```python
data2['BMI'] = data2['weight(kg)'] / (data2['height(m)'] * data2['height(m)'])
```

```python
data2.head()
```

```python
def bmi_category(x):
    if x >= 25:
        return '초고도 비만'
    
    elif x >= 23:
        return '고도 비만'
    
    elif x >= 18.5:
        return '정상'
    
    else:
        return '저체중'
```

```python
data2['BMI_Category'] = data2['BMI'].map(bmi_category)
```

```python
data2.head()
```

```python
print(data2['BMI_Category'][(data2['BMI_Category'] == '초고도 비만') | (data2['BMI_Category'] == '저체중') ].count())
```

순유입인원은 초중고 도내,도외 전입인원에서 초중고 도내, 도외 전출인원을 뺀값이다. 각년도별로 가장 큰 순유입인원을 가진 지역구의 순유입인원을 구하고 전체 기간의 해당 순유입인원들의 합을 구하여라

```python
df['순유입인원'] = \
df[[x for x in df.columns if '전입' in x]].sum(axis=1) -\
df[[x for x in df.columns if '전출' in x ]].sum(axis=1)

result = df.groupby('년도')['순유입인원'].max().sum().sum()
print(result)
```
