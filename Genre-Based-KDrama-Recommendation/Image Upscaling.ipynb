```python
from PIL import Image
import os
```

```python
def upscale_images_in_folder(input_folder, output_folder, scale_factor):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        if filename.endswith(".jpg"):
            input_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, filename)
            
            try:
                img = Image.open(input_path)
                width, height = img.size
                new_width = int(width * scale_factor)
                new_height = int(height * scale_factor)
                upscaled_img = img.resize((new_width, new_height), Image.BICUBIC)
                upscaled_img.save(output_path, "JPEG", quality=95)
                print(f"{filename} 업스케일링 완료")
            except Exception as e:
                print(f"{filename} 처리 중 오류 발생: {e}")

if __name__ == "__main__":
    input_folder = "downloaded_images"
    output_folder = "upscaled_images"
    scale_factor = 8  # 업스케일링 배율

    upscale_images_in_folder(input_folder, output_folder, scale_factor)
```
